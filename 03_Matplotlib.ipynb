# Matplotlib

```python
# importamos la librerías
import matplotlib.pyplot as plt
import numpy as np
```

# Ploteo Lineal

```python
# estética
plt.style.use('classic')
```

```python
# puntos
x = np.linspace(0,10,100)
```

$$f(x)=sen(x)$$

```python
#ploteo seno
plt.plot(x, np.sin(x))
plt.show()
```

## Dos gráficos en uno

```python
# con algunas variantes
plt.figure()
#ploteo con variantes
plt.plot(x, np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

## Dos gráficos separados

```python
# Subplot
plt.figure()

# 1ro
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))

# 2do
plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
```

```python
# Ploteo orientado a objetos
fig,ax=plt.subplots(2,1)

# Call plot() method
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
# Subplot
plt.figure()

# 1ro
plt.subplot(1,2,1)
plt.plot(x, np.sin(x))

# 2do
plt.subplot(1,2,2)
plt.plot(x, np.cos(x))
```

## Línea de senos con Style

```python
# Otro style
plt.style.use('seaborn-whitegrid')
```

```python
# color específico por nombre
plt.figure()
plt.plot(x, np.sin(x-0), 'g-.')
plt.plot(x, np.sin(x-1), 'b--')
plt.plot(x, np.sin(x-2), 'r-')
plt.plot(x, np.sin(x-3), 'k:')
plt.show()
```

## Ploteo con leyendas

```python
#nueva data
x = np.linspace(0, 10, 1000)
y = np.sin(x[:, np.newaxis] + np.pi * np.arange(0, 2, 0.5))
```

```python
# first
plt.plot(x,y[:,0],'r:',label='firts')
# second
plt.plot(x,y[:,1],'g--',label='second')

plt.legend(framealpha=1,frameon=True)

plt.xlabel('time')
plt.ylabel('sen(x)')
plt.title('Líneas de senos')
plt.show()
```

## Zoom

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
# figure
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) 

# Figure Axes 1
axes1.plot(x,y,'b')
axes1.set_xlabel('time')
axes1.set_ylabel('y')
axes1.set_title('Axes 1 title')

# Figure Axes 2
axes2.plot(y,x,'r')
axes2.set_xlabel('time')
axes2.set_ylabel('x')
axes2.set_title('Axes 2 title')

plt.show()
```

## Simple scatter plots

```python
#nueva data
x = np.linspace(0, 10, 30)
y = np.sin(x)
```

```python
# scatter
plt.scatter(x,y, marker='+')
#limitante
plt.ylim(-1.2 , 1.2)
plt.xlabel('time')
plt.ylabel('sen(x)')
plt.title('Líneas de senos')
plt.show()
```

```python
# nueva data
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
```

## Error Bars

```python
# nueva data
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
```

```python
# ploteo error
plt.errorbar(x,y, yerr=dy, fmt='.k')
```

```python
plt.errorbar(x,y, yerr=dy, fmt='o', color='black',ecolor='lightgray', elinewidth=3, capsize=0)
plt.xlabel('Points')
plt.ylabel('Tic')
plt.title('Error')
plt.show()
```

## Countour Plots

$$f(x,y)=sen(x)^{10}+cos(10+xy)\times cos(x)$$

```python
#función f
def f(x, y):
    return np.sin(x)**10 + np.cos(10+x*y)*np.cos(x)
```

```python
# nueva data
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
# contour
plt.contour(X,Y,Z, colors='black')
```

```python
# contour + cmap
plt.contour(X,Y,Z, 20, cmap='RdGy')
```

```python
# contourf
plt.contourf(X,Y,Z, 20, cmap='plasma')
# color bar
plt.colorbar()
```

## Histogramas

```python
# nueva data
data = np.random.randn(1000)
```

```python
# hist
plt.hist(data);
```

```python
# hist + variables
plt.hist(data,bins=30, density=True, alpha=0.5, histtype='stepfilled', 
         color='steelblue', edgecolor='none');
plt.xlabel('Points')
plt.ylabel('Tic')
plt.title('Error')
plt.show()
```

```python
# new data
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)
```

```python
#multiple hist
kwargs=dict(bins=40, density=True, alpha=0.3, histtype='stepfilled')
#ploteo hist
plt.hist(x1, **kwargs, label='h1')
plt.hist(x2, **kwargs, label='h2')
plt.hist(x3, **kwargs, label='h3');

plt.legend()

plt.xlabel('Points')
plt.ylabel('Tic')
plt.title('Error')
plt.show()
```

## Boxplot

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
```

```python
# box plot
plt.boxplot(data, vert=True, patch_artist=True);

plt.xlabel('Points')
plt.ylabel('Tic')
plt.title('Error')
plt.show()
```

# Customizing Matplotlib

### Ejemplo 1

```python
def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1,2, figsize=(10,4))
    ax[0].hist(np.random.randn(1000), alpha=0.3)
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend('abc', loc='lower left')
```

```python
hist_and_lines()
```

```python
with plt.style.context('fivethirtyeight'):
    hist_and_lines()
```
